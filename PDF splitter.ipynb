```python
import os
import re

# These libraries are used to work with PDFs
import PyPDF2
import io

# Open the PDF file in read-binary mode
with open('invoices (1).pdf', 'rb') as file:
    # Create a PDF object
    pdf = PyPDF2.PdfReader(file)
    idx = 0
    # Iterate over each page
    for page_number in range(len(pdf.pages)):
        # Get the page object
        page = pdf.pages[page_number]

        # Extract the text from the page
        page_text = page.extract_text()

        # Use a regular expression to search for the "Issue Date" string
        match = re.search(r'Issue Date: (\w{3} \d{2}, \d{4})', page_text)
        
        if match:
            idx = idx+1
            # If the regular expression found a match, extract the date
            date = match.group(1)
            # Use the date as the filename for the output PDF
            output_filename = f'{date}.pdf'
            print(f'Issue date: {date}')

            # Create a PDF writer object
            pdf_writer = PyPDF2.PdfWriter()

            # Add the page to the PDF writer object
            pdf_writer.add_page(page)

            # Open a new PDF file in write-binary mode
            with open(output_filename, 'wb') as output_file:
                # Write the PDF writer object to the output file
                pdf_writer.write(output_file)
        else:
            idx = idx+1
            # Use the date as the filename for the output PDF
            output_filename = f'{idx}_notfound.pdf'
            print(f'{idx}')

            # Create a PDF writer object
            pdf_writer = PyPDF2.PdfWriter()

            # Add the page to the PDF writer object
            pdf_writer.add_page(page)

            # Open a new PDF file in write-binary mode
            with open(output_filename, 'wb') as output_file:
                # Write the PDF writer object to the output file
                pdf_writer.write(output_file)
            # If the regular expression didn't find a match, print an error message
            print('Error: Could not find "Issue Date" on page')
```
